```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
NG = nx.Graph()

# Add nodes with attributes
NG.add_node(1, label="Node 1", color='blue')
NG.add_node(2, label="Node 2", color='green')
NG.add_node(3, label="Node 3", color='red')

# Add edges with weights
NG.add_edge(1, 2, weight=4.5)
NG.add_edge(2, 3, weight=3.0)
NG.add_edge(3, 1, weight=5.2)

# Get attributes
NL = nk.get_node_attributes(NG, 'label')
NC = [NG.nodes[node]['color'] for node in NG.nodes()]
EW = nk.get_edge_attributes(NG, 'weight')

# Draw the graph
pos = nk.spring_layout(NG)  # positions for all nodes
nx.draw(NG, pos, with_labels=True, node_color= NC)
nx.draw_networkx_labels(NG, pos, labels= NL)
nx.draw_networkx_edge_labels(NG, pos, edge_labels= EW)
plt.title("Network Graph with Attributes")
plt.show()
```
